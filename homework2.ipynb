```python
#concat ile Veri seti birleştirildi 
connect_data_frame = pd.concat([infected_data_frame, recovered_data_frame, deaths_data_frame], keys=["Infected", "Recovered", "Deaths"])
print(connect_data_frame)
```

```python
# Province/State sütunundaki None değerler fillna ile doldurulma
connect_data_frame["Province/State"].fillna("--", inplace=True)# inplace false ise orjinal DataFrame değişmeden kalır ve değişiklikler yeni bir DataFrame'e uygulanır
print(connect_data_frame)
```

```python
connect_data_grouped=connect_data_frame.groupby('Province/State')
for name, group in connect_data_grouped:
  print(group)
```
