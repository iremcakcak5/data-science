```python
def make_frame(country:str):
    """
    Take country name as parameter, return infected ,recovered,deaths values as timeseries
    her ülke için her dataframe  getirecek
    pd.DataFrame()
    """
    df = pd.DataFrame({
       'infected':df_infected.loc[country],
       'recovered':df_recovered.loc[country],
       'deaths':df_deaths.loc[country]
 })
      #indexler saat olması için 
    df.index =pd.to_datetime(df.index,format='%m/%d/%y')

    return df
```

```python
df_us=make_frame('US')
df_turkey=make_frame('Turkey'
                     )
```

```python
#the basic reproduction number
#yeni hasta olanlara ulaşıp 
df_us['Rt']= df_us['new_infected'].rolling(8).apply(lambda x: x[4:].sum()/x[:4].sum())
#8 gün iç.n hesaplama
```

```python
df_us
```

```python
df_us['Rt'].plot()
plt.show()
```

```python
#nan ve inf değerler için neler yapabiliriz:
#log değerler
#virgülden sonra basamak sayısını azaltmak
#eksik verileri tamamlamak
#eksik veriyi doldurmak için pandas metodlarına bakmak
```

```python
#df_us['Rt'].replace(np.inf,np.nan).fiillna('bfill')
#inf olan tanımsız olan değerleri na ile doldurduk
df_us['Rt'].replace(np.inf, np.nan).fillna(method='bfill')
```
