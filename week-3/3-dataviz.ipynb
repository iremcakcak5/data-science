```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches
#pip install seaborn
import pandas as pd
```

```python
df= pd.DataFrame({
    'Group':['A','B','C','D','E'],
    'Value':[1,5,4,3,9]
})
```

```python
plt.barh(y=df.Group, width=df.Value)
```

```python

df = df.sort_values(by=['Value'])
my_plot = plt.barh(
    y=df.Group,
    width=df.Value,
    color=mcolors.TABLEAU_COLORS,
    edgecolor='black',
    linewidth=3
    )
plt.bar_label(container=my_plot, padding=4)
plt.show()
#barh metodunda sÄ±ralama
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.set(style="darkgrid")
sns.barplot(
    x="total_bill",
    y="day",
    data=tips,
    color='cyan'
)
plt.figure(figsize=(18, 6))
```

```python
tips.head()
```

```python
total=tips.groupby('day')['total_bill'].sum().reset_index()
```

```python
total
```

```python
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='darkblue')
```

```python
smoker = tips[tips.smoker=='Yes']
```

```python
bar2 = sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)
```

```python
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='darkblue')
bar2 = sns.barplot(
    x='day',
    y='total_bill',
    data=smoker,
    estimator=sum,
    errorbar=None,
    color='lightblue'
)
top_bar = mpatches.Patch(color='darkblue', label='non-smoker')
bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes')
plt.legend(handles=[top_bar, bottom_bar])
plt.show()
```

```python
sns.barplot(
        x="total_bill",
        y="smoker",
        hue="sex",
        data=tips,
        palette="pink"
        )
plt.figure(figsize=(16, 8))
```

```python
df_iris=sns.load_dataset('iris')
```

```python
df_iris.head()
```

```python
sns.histplot(data= df_iris, x='sepal_length')
```

```python
sns.distplot(df_iris["sepal_length"])
#distplot is deprecated
```

```python
sns.regplot(x=df_iris['sepal_length'],y=df_iris['sepal_width'])
```

```python
sns.regplot(
    x=df_iris['sepal_length'],
    y=df_iris['sepal_width'],
    marker="+",
    line_kws={
        "color":"red",
        "alpha":0.6,
        "lw":5
    },
    scatter_kws={
        "color":"darkblue",
        "s":50
    }
    )
```

```python
sns.lmplot(
    x='sepal_length',
    y='sepal_width',
    data=df_iris,
    hue='species'
)
```

```python
sns.set_theme(style="white", palette="pastel")
sns.boxplot(x=df_iris['species'], y=df_iris['sepal_length'])
```
